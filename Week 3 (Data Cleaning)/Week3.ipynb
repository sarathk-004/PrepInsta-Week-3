# **Dataset 1 - Data Cleaning using Pandas**

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
mean_age = data['Age'].mean()
data['Age'].fillna(mean_age, inplace=True)
data['Rating'].replace(-1, 0, inplace=True)
data['Rating'].replace('', 0, inplace=True)
```

```python
data
```

```python
data['Salary'] = data['Salary'].str.replace('$', '').str.replace('k', "000")
data['Established'] = pd.to_datetime(data['Established'], errors='coerce')
data['Easy Apply'] = data['Easy Apply'].replace('-1', False)
```

```python
data
```

```python
Q1 = data[['Age', 'Salary', 'Rating','Location','Established','Easy Apply']].quantile(0.75)
Q3 = data[['Age', 'Salary', 'Rating','Location','Established','Easy Apply']].quantile(0.25)
A = Q3 - Q1

threshold = 1

outliers = ((data[['Age', 'Salary', 'Rating']] < (Q1 - threshold * A)) | (data[['Age', 'Salary', 'Rating']] > (Q3 + threshold * A))).any(axis=1)
```

```python
data
```

```python
# Assuming df is the DataFrame containing your data
data['Salary_Avg'] = data['Salary'].apply(lambda x: np.mean([int(s) for s in x.split('-')]))
```

```python
data
```

```python
data['Location'] = data['Location'].str.replace(',', ', ')
data['Location'] = data['Location'].str.replace(', In', '', regex=True).str.replace(', Ny', '', regex=True).str.replace(', Aus', '', regex=True)
data['Location'] = data['Location'].str.replace(' In', '', regex=True).str.replace(' Aus', '', regex=True)
```

```python
data
```

```python
data['Established'] = pd.to_datetime(data['Established'])

data['Established'] = data['Established'].dt.strftime('%Y-%m-%d')
```

```python
data
```

```python
data['Easy Apply'] = data['Easy Apply'].replace({'TRUE': True})
```

```python
data
```

```python
print("Unique Ratings:", data['Rating'].unique())

rating_stats = data['Rating'].describe()
print("\nSummary Statistics for Rating:\n", rating_stats)
data['Rating'].fillna(0, inplace=True)
```

```python
data
```

```python
data['Location'] = data['Location'].str.replace('[^a-zA-Z0-9\s]', '')
```

```python
data['Age'] = data['Age'].round()
```

```python
data
```

```python
duplicates = data[data.duplicated()]
unique_locations = data['Location'].unique()
unique_easy_apply = data['Easy Apply'].unique()
descriptive_stats = data.describe()
data['Established'] = pd.to_datetime(data['Established'], errors='coerce')
inconsistent_entries = data[data['Location'] == 'New York'][data['Established'] != '1969-12-31']
unique_easy_apply = data['Easy Apply'].unique()
```

```python
data
```

```python
data['Normalized_Rating'] = (data['Rating'] - data['Rating'].min()) / (data['Rating'].max() - data['Rating'].min())

print(data[['Rating', 'Normalized_Rating']])
```

```python
data['Adjusted_Rating'] = (data['Rating'] - data['Rating'].min()) / (data['Rating'].max() - data['Rating'].min()) * 9 + 1

print(data[['Rating', 'Adjusted_Rating']])
```
